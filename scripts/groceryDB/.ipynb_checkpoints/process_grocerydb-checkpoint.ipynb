```python
from dsc80_utils import *
```

```python
cols = [
    # 'original_ID',
    'name',
    'store',
    'harmonized single category',
    'brand',
    'f_FPro',
    # 'f_FPro_P',
    # 'f_min_FPro',
    # 'f_std_FPro',
    'f_FPro_class',
    'price',
    'price percal',
    'package_weight',
    'has10_nuts',
    'is_Nuts_Converted_100g',
    'Protein',
    'Total Fat',
    'Carbohydrate',
    'Sugars, total',
    'Fiber, total dietary',
    # 'Calcium',
    # 'Iron',
    'Sodium',
    # 'Vitamin C',
    'Cholesterol',
    # 'Fatty acids, total saturated',
    # 'Total Vitamin A',
]
col_renames = {
    'harmonized single category': 'category',
    'f_FPro': 'FPro',
    'f_FPro_class': 'FPro_class',
}
```

```python
def subset_and_rename(df):
    return df[cols].rename(columns=col_renames)
```

```python
def filter_has10(df):
    return (
        df.query('has10_nuts == 1 and is_Nuts_Converted_100g == 1')
        .drop(columns=['has10_nuts', 'is_Nuts_Converted_100g'])
    )
```

```python
df = (
    pd.read_csv('GroceryDB_foods.csv')
    .pipe(subset_and_rename)
    .pipe(filter_has10)
)
df
```

```python
display_df(df.describe(), rows=8, cols=12)
```

```python
df.to_csv('../../resources/data/grocerydb.csv', index=False)
```
